```python
import pandas as pd
from bs4 import BeautifulSoup
import time
from datetime import datetime
import requests
```

```python
def Scrape_all(job,country,country_name,j_codes_hold):
    Job_list_Second = []
    country_list = country 
    c_count = 0

    for country_nn in country_list:
        count = 0
        c_count = c_count + 1
        country_df = country_name.get(c_count)

        for job_name in job:
            count = count + 1
            for i in range(0,200,10):
                url = "https://"+ country_nn + ".indeed.com/jobs?q=title%3A("+ str(job_name) + ")+-master+-phd+-ODD+-MD+-MBA+-MA+-MS&jt=fulltime&explvl=entry_level&start=" + str(i)
                r=requests.get(url)
                print(url)
                time.sleep(1)
                s=BeautifulSoup(r.text, "html.parser")
                divs = s.find_all('div', class_='job_seen_beacon')
                for job_posting in divs:
                    try:
                        title = job_posting.find('h2').text
                    except:
                        title = 'None'
                    try:   
                        salary = job_posting.find('span',class_='salary-snippet').text
                    except:
                        salary = 'None'
                    try:
                        summary = job_posting.find('div',class_='job-snippet').text.strip().replace('\n','')
                    except:
                        summary = 'None'
                    try:
                        company = job_posting.find('span',class_='companyName').text
                    except:
                        company = 'None'
                    try:
                        location = job_posting.find('div',class_='companyLocation').text
                    except:
                        location = 'None'
      

                    now = datetime.now()
                    today = now.strftime("%m/%d/%Y")
                    website = 'Indeed'
                    job_level = 'Part Time'
                    
                    job_type,certificate = j_codes_hold.get(count)
                    info = {
                        'title' : title,
                        'certificate': certificate,
                        'salary': salary,
                        'location':location,
                        'summary' : summary,
                        'company' : company,
                        'Posted_date' : today,
                        'Country_name' : country_df,
                        'Job_type' : job_type,
                        'Website' : website,
                        'Job_level' : job_level
                    }
                    Job_list_Second.append(info)
        df = pd.DataFrame(Job_list_Second)
        df = df.drop_duplicates(subset=['title','salary','location','summary','company','certificate'])
        df.to_csv('neew.csv', mode='a', index = False, header=None)
```
