```python
def Scrape1(country_list,country_name_list,j_codes,job_name_list):
    Job_list_Second = []
    c_count = 0
    for country_url in country_list:
        count = 0
        c_count = c_count + 1
        country_fullname = country_name_list.get(c_count)
        
        for job_name in job_name_list:
            url = "https://"+ str(country_url) + ".indeed.com/jobs?q=\""+job_name +"\"+-master+-phd+-ODD+-MD+-MBA+-MA+-MS&jt&fromage=1"
            time.sleep(10)
            r=requests.get(url)
            s=BeautifulSoup(r.text, "html.parser")
            print(url)  
            
            count = count + 1
            try:
                count_job = s.find("div", {"id": "searchCountPages"}).find(text=True)
            except:
                count_job = '0'
                
            now = datetime.now()
            today = now.strftime("%m/%d/%Y")
            website = 'Indeed'
            global j_codes1
            job_type = j_codes.get(count)
                
            info = {
                'Total_job' : count_job,
                'Posted_date' : today,
                'Country_name' : country_fullname,
                'Job_type' : job_type,
                'Website' : website,
                'Title' : job_name
                }
            Job_list_Second.append(info)
            print(info)
    df2 = pd.DataFrame(Job_list_Second)
    df2['Total_job'] = df2['Total_job'].str.replace(',', '')
    df2["Replace_1"]= df2["Total_job"].str.replace(' 1 ', "NotUsing", case = False)  
    df2 = df2.assign(Total = lambda x: x['Replace_1'].str.extract('(\d+\s+)'))
    df2=df2.drop(columns=['Replace_1'])
    df2['Total'].fillna(1, inplace=True)
    df2.loc[df2['Total_job'] == '0', 'Total'] = 0
    df2.to_csv('scraping_.csv', mode='a', index = False, header=None)
```

```python
import schedule
import time

schedule.every().day.at("17:00").do(lambda:Scrape1(country_list_1,country_name_1,j_codes_1,job_eng_1))
schedule.every().day.at("19:30").do(lambda:Scrape1(country_list_1,country_name_1,j_codes_2,job_eng_2))
schedule.every().day.at("20:30").do(lambda:Scrape1(country_list_1,country_name_1,j_codes_3,job_eng_3))
schedule.every().day.at("21:30").do(lambda:Scrape1(country_list_1,country_name_1,j_codes_4,job_eng_4))
schedule.every().day.at("22:30").do(lambda:Scrape1(country_list_1,country_name_1,j_codes_5,job_eng_5))

while True:
    schedule.run_pending()
    time.sleep(1)
```
